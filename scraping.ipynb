```python
# Import libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException

# Configure selenium
# options = Options()
# options.headless = True
# options.add_argument("--window-size=1920,1200")
driver = webdriver.Chrome()
driver.get("https://www.tokopedia.com/search?navsource=&ob=5&srp_component_id=04.06.00.00&srp_page_id=&srp_page_title=&st=&q=laptop")

time.sleep(10)

# Scraping process
products = driver.find_elements(By.XPATH, "//div[@class='prd_link-product-name css-3um8ox']")

# Create lists for data
data = []

for product in products:
    product_name = product.text

    try:
        price_element = product.find_element(By.XPATH, ".//following-sibling::div[@class='prd_link-product-price css-1ksb19c']")
        price = price_element.text
    except NoSuchElementException:
        price = None

    try:
        location_element = product.find_element(By.XPATH, ".//following-sibling::span[@class='prd_link-shop-loc css-1kdc32b flip']")
        location = location_element.text
    except NoSuchElementException:
        location = None

    try:
        rating_element = product.find_element(By.XPATH, ".//following-sibling::span[@class='prd_rating-average-text css-t70v7i']")
        rating = rating_element.text
    except NoSuchElementException:
        rating = None

    try:
        sales_element = product.find_element(By.XPATH, ".//following-sibling::span[@class='prd_label-integrity css-1duhs3e']")
        sales = sales_element.text
    except NoSuchElementException:
        sales = None

    data.append({
        'Product': product_name,
        'Price': price,
        'Rating': rating,
        'Location': location,
        'Sales': sales,
    })

# Create a dataframe from the collected data
df = pd.DataFrame(data)
print(df)

driver.quit()
```

```python
# Import libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException

# Configure selenium
driver = webdriver.Chrome()
driver.get("https://www.tokopedia.com/search?navsource=&ob=5&srp_component_id=04.06.00.00&srp_page_id=&srp_page_title=&st=&q=laptop")

counter_page = 0

while counter_page < 5:
    for _ in range(0, 6500, 500):
        time.sleep(0.1)
        driver.execute_script("window.scrollBy(0,500)")

    time.sleep(10)

    # Scraping process
    products = driver.find_elements(By.XPATH, "//div[@class='prd_link-product-name css-3um8ox']")

    # Create lists for data
    data = []

    for product in products:
        product_name = product.text

        parent = product.find_element(By.XPATH, "..")

        try:
            price_element = parent.find_element(By.XPATH, ".//div[@class='prd_link-product-price css-1ksb19c']")
            price = price_element.text
        except NoSuchElementException:
            price = None

        try:
            location_element = parent.find_element(By.XPATH, ".//span[@class='prd_link-shop-loc css-1kdc32b flip']")
            location = location_element.text
        except NoSuchElementException:
            location = None

        try:
            rating_element = parent.find_element(By.XPATH, ".//span[@class='prd_rating-average-text css-t70v7i']")
            rating = rating_element.text
        except NoSuchElementException:
            rating = None

        try:
            sales_element = parent.find_element(By.XPATH, ".//span[@class='prd_label-integrity css-1duhs3e']")
            sales = sales_element.text
        except NoSuchElementException:
            sales = None

        data.append({
            'Product': product_name,
            'Price': price,
            'Rating': rating,
            'Location': location,
            'Sales': sales,
        })

        counter_page += 1

        next_page = driver.find_element(by=By.XPATH, value='//*[@id="zeus-root"]/div/div[2]/div/div[2]/div[5]/nav/ul/li[11]/button')
        next_page.click()

# Create a dataframe from the collected data
df = pd.DataFrame(data)
print(df)

driver.quit()
```
